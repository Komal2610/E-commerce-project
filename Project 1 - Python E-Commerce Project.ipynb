```python
# Question 5
# Find the total Revenue generated by the E-Commerce company over all orders placed. Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total revenue.
```

```python
#function to determine the revenue
def compute_rev(x):
    if x > 600:
        return x*0.25
    elif x > 350:
        return x*0.15
    elif x > 100:
        return x*0.10
    else:
        return x*0.05

df['Revenue'] = df['discounted_price'].apply(compute_rev)
df.head()
```

```python
total_rev = df['Revenue'].sum()
```

```python
# Question 6
# Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order .Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue. Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)
```

```python

# Calculate BrandRevenue
df['BrandRevenue'] = df['discounted_price'] - df['Revenue']

# Group by brand and calculate total BrandRevenue
brand_revenue_df = df.groupby('Brand')['BrandRevenue'].sum().reset_index()

# Sort the DataFrame by total BrandRevenue in descending order
sorted_brand_revenue_df = brand_revenue_df.sort_values(by='BrandRevenue', ascending=False)

# List the top 10 brands
top_10_brands = sorted_brand_revenue_df.head(10)

print(top_10_brands)
```

```python
# Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately. 
# Identify the best and the worst performing months for each region. 
```

```python

df['Delivered_Date'] = pd.to_datetime(df['Delivered_Date'])

# Plotting the monthly revenue for each region separately
regions = df['Region'].unique()

plt.figure(figsize=(12, 8))

for region in regions:
    region_data = df[df['Region'] == region]
    monthly_revenue = region_data.resample('M', on='Delivered_Date')['Revenue'].sum()
    
    plt.plot(monthly_revenue, label=region)

plt.title('Monthly Revenue of E-Commerce Company by Region')
plt.xlabel('Delivered Date')
plt.ylabel('Revenue')
plt.legend(loc='upper left')
plt.show()
```

```python

df['Delivered_Date'] = pd.to_datetime(df['Delivered_Date'])

# Extracting the month and year from the 'Delivered_Date' column
df['YearMonth'] = df['Delivered_Date'].dt.to_period('M')

# Group by 'Region' and 'YearMonth', calculate the total revenue for each group
region_monthly_revenue = df.groupby(['Region', 'YearMonth'])['Revenue'].sum().reset_index()

# Identify the best and worst performing months for each region
best_months = region_monthly_revenue.loc[region_monthly_revenue.groupby('Region')['Revenue'].idxmax()]
worst_months = region_monthly_revenue.loc[region_monthly_revenue.groupby('Region')['Revenue'].idxmin()]

print("Best Performing Months:")
print(best_months)

print("\nWorst Performing Months:")
print(worst_months)
```
